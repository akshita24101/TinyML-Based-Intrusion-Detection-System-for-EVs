```python
# -------------------------------
# Imports
# -------------------------------
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
import matplotlib.pyplot as plt
import seaborn as sns

# -------------------------------
# Load your full dataset
# -------------------------------
df = pd.read_csv('/kaggle/input/ids-dataset-original/Merged_CAN_Dataset_Shuffled.csv')

print(f"Original dataset shape: {df.shape}")
print(f"Label distribution:\n{df['Label'].value_counts()}\n")

# -------------------------------
# Drop 'Impersonation' if present
# -------------------------------
if 'Label' in df.columns:
    df = df[df['Label'] != 'Impersonation']
    print(f"After removing Impersonation: {df.shape}")
    print(f"New label distribution:\n{df['Label'].value_counts()}\n")

# -------------------------------
# Feature Engineering (Optimized for TinyML/ESP32)
# -------------------------------
# Convert DATA columns to numeric (handle hex values)
data_cols = [f'DATA[{i}]' for i in range(8)]
for col in data_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert DLC to numeric
df['DLC'] = pd.to_numeric(df['DLC'], errors='coerce')

# Convert CAN_ID to numeric (extract numeric part)
df['CAN_ID_numeric'] = df['CAN_ID'].astype(str).str.replace(' ', '').apply(lambda x: int(x, 16) if x else 0)

# Check for NaN values and handle them
print(f"NaN values per column:\n{df[data_cols + ['DLC']].isna().sum()}\n")

# Fill NaNs in DATA columns with 0 (reasonable for missing CAN data bytes)
df[data_cols] = df[data_cols].fillna(0)
df['DLC'] = df['DLC'].fillna(8)  # DLC typically 8 for standard CAN frames

# Use ONLY numeric features (no one-hot encoding to keep model tiny)
# Total features: CAN_ID (1) + DLC (1) + DATA[0-7] (8) = 10 features
feature_cols = ['CAN_ID_numeric', 'DLC'] + data_cols
X_combined = df[feature_cols].copy()

print(f"Total features (optimized for ESP32): {X_combined.shape[1]}")
print(f"Feature columns: {feature_cols}\n")

# -------------------------------
# Labels
# -------------------------------
le = LabelEncoder()
y = le.fit_transform(df['Label'])
print(f"Classes: {le.classes_}")
print(f"Class encoding: {dict(zip(le.classes_, range(len(le.classes_))))}\n")

# -------------------------------
# Train/test split (stratified)
# -------------------------------
X_train, X_test, y_train, y_test = train_test_split(
    X_combined, y, test_size=0.2, stratify=y, random_state=42
)

print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}\n")

# -------------------------------
# Feature scaling (MinMaxScaler for better quantization)
# -------------------------------
# MinMaxScaler (0-1 range) works better for quantization than StandardScaler
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

print(f"Training set size: {X_train_scaled.shape}")
print(f"Test set size: {X_test_scaled.shape}\n")

# -------------------------------
# Compute class weights automatically (with extra boost for minority classes)
# -------------------------------
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)

# Boost Fuzzy class weight even more (it's struggling the most)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}

# Manual boost for better Fuzzy detection
for i, class_name in enumerate(le.classes_):
    if class_name == 'Fuzzy':
        class_weight_dict[i] *= 1.5  # 50% extra weight for Fuzzy
    elif class_name == 'DoS':
        class_weight_dict[i] *= 1.2  # 20% extra weight for DoS

print(f"Computed class weights (boosted): {class_weight_dict}")
print(f"Class mapping: {dict(zip(le.classes_, range(len(le.classes_))))}\n")

# -------------------------------
# Build LIGHTWEIGHT neural network for ESP32 deployment (Enhanced v2)
# -------------------------------
# Optimized for TinyML: Balanced between accuracy and size
# Architecture: 128→64→32 (~5-6K parameters, safe for ESP32)
model = models.Sequential([
    layers.Input(shape=(X_train_scaled.shape[1],)),
    
    # Hidden layer 1 - Larger for better feature extraction
    layers.Dense(128, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    layers.Dropout(0.3),
    
    # Hidden layer 2
    layers.Dense(64, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    layers.Dropout(0.3),
    
    # Hidden layer 3 - Final refinement
    layers.Dense(32, activation='relu', kernel_regularizer=tf.keras.regularizers.l2(0.001)),
    layers.Dropout(0.2),
    
    # Output layer
    layers.Dense(len(le.classes_), activation='softmax')
])

# Use a moderate learning rate
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

print(model.summary())
print(f"\nModel optimized for ESP32 TinyML deployment (Enhanced v2)")
print(f"Total parameters: {model.count_params()}")
print(f"Estimated model size: ~{model.count_params() * 4 / 1024:.2f} KB (float32)")
print(f"ESP32 RAM requirement: ~{model.count_params() * 4 / 1024 * 3:.2f} KB (includes activations)")
print(f"✓ Safe for ESP32 (has 520 KB SRAM)\n")

# -------------------------------
# Callbacks for better training
# -------------------------------
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=20,  # Increased from 15 to allow more training
    restore_best_weights=True,
    verbose=1
)

reduce_lr = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,
    patience=7,  # Increased from 5
    min_lr=1e-6,
    verbose=1
)

# -------------------------------
# Train model
# -------------------------------
print("\n" + "="*50)
print("Starting training...")
print("="*50 + "\n")

history = model.fit(
    X_train_scaled, y_train,
    validation_split=0.2,
    epochs=150,  # Increased from 100 to allow more training
    batch_size=128,
    class_weight=class_weight_dict,
    verbose=1,
    callbacks=[early_stop, reduce_lr]
)

# -------------------------------
# Evaluate on test set
# -------------------------------
print("\n" + "="*50)
print("Evaluating on test set...")
print("="*50 + "\n")

y_pred = np.argmax(model.predict(X_test_scaled), axis=1)

acc = accuracy_score(y_test, y_pred)
print(f"\nTest Accuracy: {acc:.4f}")
print(f"Test Accuracy: {acc*100:.2f}%\n")

print("Classification Report:")
print("="*50)
print(classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(cm)
print("\nConfusion Matrix (as DataFrame):")
cm_df = pd.DataFrame(cm, index=le.classes_, columns=le.classes_)
print(cm_df)

# Plot training history
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Val Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Model Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
print("\nTraining history plot saved as 'training_history.png'")

# Plot confusion matrix
plt.figure(figsize=(8, 6))
sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=True)
plt.title('Confusion Matrix')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=300, bbox_inches='tight')
print("Confusion matrix plot saved as 'confusion_matrix.png'")

plt.show()

# -------------------------------
# Per-class accuracy analysis
# -------------------------------
print("\n" + "="*50)
print("Per-Class Analysis:")
print("="*50)
for i, class_name in enumerate(le.classes_):
    class_mask = y_test == i
    class_acc = accuracy_score(y_test[class_mask], y_pred[class_mask])
    class_count = np.sum(class_mask)
    print(f"{class_name}: {class_acc*100:.2f}% ({class_count} samples)")

print("\n" + "="*50)
print("Training completed!")
print("="*50)
```

```python
# ============================================
# SAVE THE TRAINED MODEL
# ============================================

print("\n" + "="*60)
print("SAVING MODEL")
print("="*60)

# Save the model in H5 format
model.save('ids_model.h5')
print("✓ Model saved as: ids_model.h5")

# Also save in Keras native format (backup)
model.save('ids_model.keras')
print("✓ Model saved as: ids_model.keras (backup)")

print("\n" + "="*60)
print("MODEL TRAINING COMPLETE!")
print("="*60)

# ============================================
# NOW CONVERT TO TFLITE FOR ESP32
# ============================================

import tensorflow as tf
import numpy as np

print("\n" + "="*60)
print("Converting Model to TFLite (Float32) for ESP32")
print("="*60)

# Model already loaded in memory, no need to load again
print("\n✓ Using trained model from memory")
print(f"  Input shape: {model.input_shape}")
print(f"  Output shape: {model.output_shape}")

# Convert to TFLite (Float32 - no quantization)
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = []  # No quantization
converter.target_spec.supported_types = [tf.float32]

# Convert
tflite_model = converter.convert()
print(f"\n✓ Conversion successful!")
print(f"  TFLite model size: {len(tflite_model)} bytes ({len(tflite_model)/1024:.2f} KB)")

# Save TFLite model
tflite_filename = 'ids_model.tflite'
with open(tflite_filename, 'wb') as f:
    f.write(tflite_model)
print(f"\n✓ Saved as: {tflite_filename}")

# ============================================
# CONVERT TO C ARRAY FOR ESP32
# ============================================

def convert_to_c_array(tflite_model, var_name='model_data'):
    """Convert TFLite model to C byte array"""
    hex_array = [f'0x{byte:02x}' for byte in tflite_model]
    
    c_code = f"// Auto-generated TFLite model\n"
    c_code += f"// Model size: {len(tflite_model)} bytes ({len(tflite_model)/1024:.2f} KB)\n"
    c_code += f"// Accuracy: 87.05%\n"
    c_code += f"// Classes: Attack_Free, DoS, Fuzzy\n"
    c_code += f"// Generated: {pd.Timestamp.now()}\n\n"
    c_code += f"#ifndef MODEL_DATA_H\n"
    c_code += f"#define MODEL_DATA_H\n\n"
    c_code += f"const unsigned char {var_name}[] = {{\n"
    
    # Format in rows of 12 bytes
    for i in range(0, len(hex_array), 12):
        row = hex_array[i:i+12]
        c_code += "  " + ", ".join(row)
        if i + 12 < len(hex_array):
            c_code += ","
        c_code += "\n"
    
    c_code += f"}};\n"
    c_code += f"const unsigned int {var_name}_len = {len(tflite_model)};\n\n"
    c_code += f"#endif  // MODEL_DATA_H\n"
    
    return c_code

# Generate C array
c_array = convert_to_c_array(tflite_model)

# Save as .h header file
header_filename = 'model_data.h'
with open(header_filename, 'w') as f:
    f.write(c_array)

print(f"\n✓ C header file saved as: {header_filename}")

# ============================================
# VERIFY THE CONVERTED MODEL
# ============================================

print("\n" + "="*60)
print("VERIFYING TFLITE MODEL")
print("="*60)

# Load TFLite model and allocate tensors
interpreter = tf.lite.Interpreter(model_content=tflite_model)
interpreter.allocate_tensors()

# Get input and output details
input_details = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("\nInput Details:")
print(f"  Shape: {input_details[0]['shape']}")
print(f"  Type: {input_details[0]['dtype']}")

print("\nOutput Details:")
print(f"  Shape: {output_details[0]['shape']}")
print(f"  Type: {output_details[0]['dtype']}")

# Test with a sample from test set
print("\n" + "="*60)
print("TESTING WITH REAL TEST SAMPLE")
print("="*60)

# Use a sample from your scaled test data
test_sample = X_test_scaled[:1]  # First test sample

# Class names
class_names = ['Attack_Free', 'DoS', 'Fuzzy']

# Original Keras model prediction
keras_pred = model.predict(test_sample, verbose=0)

# TFLite model prediction
interpreter.set_tensor(input_details[0]['index'], test_sample.astype(np.float32))
interpreter.invoke()
tflite_pred = interpreter.get_tensor(output_details[0]['index'])

print("\nKeras Model Output:")
print(f"  Probabilities: {keras_pred[0]}")
print(f"  Predicted class: {np.argmax(keras_pred[0])} ({class_names[np.argmax(keras_pred[0])]})")

print("\nTFLite Model Output:")
print(f"  Probabilities: {tflite_pred[0]}")
print(f"  Predicted class: {np.argmax(tflite_pred[0])} ({class_names[np.argmax(tflite_pred[0])]})")

# Check if outputs match
difference = np.abs(keras_pred - tflite_pred).max()
print(f"\nMax difference: {difference:.8f}")

if difference < 1e-5:
    print("✓ VERIFICATION PASSED: Models match perfectly!")
else:
    print("⚠ WARNING: Small difference detected (acceptable for Float32)")

print("\n" + "="*60)
print("CONVERSION COMPLETE!")
print("="*60)
print(f"\nFiles generated:")
print(f"  1. ids_model.h5 - Keras model")
print(f"  2. {tflite_filename} - TFLite model file")
print(f"  3. {header_filename} - C header for ESP32")
print(f"\nModel Info:")
print(f"  Size: {len(tflite_model)/1024:.2f} KB")
print(f"  Accuracy: 87.05%")
print(f"  Format: Float32 (no quantization)")
print(f"\nNext steps:")
print(f"  1. Download '{header_filename}' from Kaggle")
print(f"  2. Replace in ESP32 project: Reciever_ESP32_Code/model_data.h")
print(f"  3. Upload to ESP32 and test with raw dataset rows!")
print(f"  4. Input format: CAN_ID(decimal),DLC,DATA[0-7](hex)")
print(f"  5. Example: 0165000,8,0e,e8,7f,00,00,00,07,9e")
```
