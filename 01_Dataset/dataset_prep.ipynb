Simple Merge (Labels together)

```python
import pandas as pd

# Load all datasets
dos = pd.read_csv("DoS_attack_dataset.csv")
fuzzy = pd.read_csv("Fuzzy_attack_dataset.csv")
impersonation = pd.read_csv("Impersonation_attack_dataset.csv")
attack_free = pd.read_csv("Attack_free_dataset.csv")

# Drop Timestamp column
for df in [dos, fuzzy, impersonation, attack_free]:
    if 'Timestamp' in df.columns:
        df.drop(columns=['Timestamp'], inplace=True)

# Add labels
dos['Label'] = 'DoS'
fuzzy['Label'] = 'Fuzzy'
impersonation['Label'] = 'Impersonation'
attack_free['Label'] = 'Attack_Free'

# Merge them into one dataset
merged = pd.concat([dos, fuzzy, impersonation, attack_free], ignore_index=True)

# Save to CSV
merged.to_csv("Merged_CAN_Dataset.csv", index=False)

print("✅ Merged dataset saved as 'Merged_CAN_Dataset.csv'")
print("Shape:", merged.shape)
print("Class distribution:\n", merged['Label'].value_counts())
```

Shuffled dataset (labels not together)

```python
import pandas as pd
# Load merged dataset
merged_path = "Merged_CAN_Dataset.csv"
df = pd.read_csv(merged_path)

# Shuffle the dataset randomly but reproducibly
df_shuffled = df.sample(frac=1, random_state=42).reset_index(drop=True)

# Save the shuffled dataset
shuffled_path = "Merged_CAN_Dataset_Shuffled.csv"
df_shuffled.to_csv(shuffled_path, index=False)

print("Shuffled dataset saved as:", shuffled_path)
print("Shape:", df_shuffled.shape)
print("Class distribution:\n", df_shuffled["Label"].value_counts())
```
